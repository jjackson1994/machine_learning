```python
# A template is provided, but feel free to build a different implementation

def perceptron_update(current_w, x, y):
    l = 0.2
    new_w = current_w + (y*l*x)
    
    return new_w


def perceptron(X, Y, max_num_iterations):
    # Place in this function the main section of the perceptron algorithm
    
    #init the algorith with w=2, use a best_w variable to keep track of the best solution
    curr_w = np.zeros(4)
    best_w = np.zeros(4)
    num_samples = len(X)
    best_error = 2
    
 #   index_misclassified = -1 #will be ovewritten
    num_classified_buono = 0 #will be ovewritten
    
    
    #main loop continue until all samples correctly classified or max # iterations reached
    num_iter = 1
    
    while ((num_classified_buono < num_samples) and (num_iter < max_num_iterations)):
        num_classified_buono = 0
        
        # avoid working always on the same sample, you can use a random permutation or randomize the choice of misclassified
        
        # PLACE YOUR CODE TO RANDOMIZE
        permutation = np.random.permutation(len(X)) # random permutation
        
        X = X[permutation]
        Y = Y[permutation]
        
        for i in range(num_samples):
            #print('x =',X[i],'w =', curr_w)
            #print('mult',curr_w*X[i])
            #print('sum',sum(curr_w*X[i]))
 
            sample_pred = int(np.sign(sum(curr_w*X[i])))
            
            #check if the i-th randomly selected sample is misclassified
            if sample_pred != Y[i]:
                curr_w = perceptron_update(curr_w, X[i], Y[i])
                #print('updated w = ',curr_w)
                
                #error calculation
                er_count = 0         
                prediction = np.sign(np.sum(curr_w.reshape(1,4)*X, axis = 1))            
                for i in range(num_samples):
                    #print(int(prediction[i]),Y[i])
                    if int(prediction[i]) != Y[i]: er_count += 1
                #print('error count',er_count)
                error = er_count / num_samples
                
                #update best w and best error if the error is better
                if error < best_error:
                    best_error = error
                    best_w = curr_w
            
            else: num_classified_buono += 1
        num_iter += 1
        
    
    return best_w, best_error
```
